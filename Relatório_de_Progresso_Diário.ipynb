# Depois de instaladas vamos importá-las

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

# Análise Preliminar dos Dados:

* Neste momento, vamos carregar os dados e verificar de forma exploratória a qualidade dos mesmos, bem como realizar uma descrição para entender os tipos de dados que serão trabalhados.

```python
# Criando o DataFrame com os dados fornecidos
data = {'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
        'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
        'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
        'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]}
df = pd.DataFrame(data)
```

# Em seguida, vamos definir duas funções:

* **show_statistics** e **show_productivity_chart**. A função show_statistics calcula e exibe as estatísticas solicitadas. A função **show_productivity_chart** cria e exibe um gráfico de barras que mostra a produtividade diária (tarefas concluídas por hora) para cada dia da semana.

```python
# Função para calcular e exibir as estatísticas solicitadas
def show_statistics(df):
    total_horas = df['Horas Trabalhadas'].sum()
    media_horas = df['Horas Trabalhadas'].mean()
    total_bugs = df['Bugs Corrigidos'].sum()
    media_bugs = df['Bugs Corrigidos'].mean()
    total_tarefas = df['Tarefas Concluídas'].sum()
    media_tarefas = df['Tarefas Concluídas'].mean()
    df['Produtividade'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

    print('Relatório de Progresso Diário')
    print('Total de Horas Trabalhadas:', total_horas)
    print('Média Diária de Horas Trabalhadas:', media_horas)
    print('Total de Bugs Corrigidos:', total_bugs)
    print('Média Diária de Bugs Corrigidos:', media_bugs)
    print('Total de Tarefas Concluídas:', total_tarefas)
    print('Média Diária de Tarefas Concluídas:', media_tarefas)
    print('Produtividade Diária (Tarefas Concluídas por Hora):')
    print(df[['Dia', 'Produtividade']])

# Função para criar um gráfico de barras para mostrar a produtividade diária
def show_productivity_chart(df):
    sns.barplot(x='Dia', y='Produtividade', data=df)
    plt.title('Produtividade Diária (Tarefas Concluídas por Hora)')
    plt.xlabel('Dia')
    plt.ylabel('Produtividade')
    plt.show()
```

# Agora podemos chamar essas funções para exibir as estatísticas e o gráfico de produtividade:

```python
# Exibindo as estatísticas e o gráfico de produtividade
show_statistics(df)
show_productivity_chart(df)
```

# Além disso, podemos realizar uma análise exploratória dos dados para obter insights adicionais.
* Por exemplo, podemos adicionar uma coluna de produtividade ao DataFrame e exibir informações estatísticas básicas sobre os dados. Em seguida, podemos criar gráficos de dispersão para mostrar a relação entre as horas trabalhadas e a produtividade e entre os bugs corrigidos e a produtividade. A partir desses gráficos, podemos observar se existe alguma correlação entre essas variáveis.

```python

# Adicionando a coluna de produtividade ao DataFrame
df['Produtividade'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

# Exibindo informações estatísticas básicas sobre o DataFrame
print('Informações Estatísticas Básicas:')
print(df.describe())

# Criando um gráfico de dispersão para mostrar a relação entre as horas trabalhadas e a produtividade
sns.scatterplot(x='Horas Trabalhadas', y='Produtividade', data=df)
plt.title('Relação entre Horas Trabalhadas e Produtividade')
plt.xlabel('Horas Trabalhadas')
plt.ylabel('Produtividade')
plt.show()

# Criando um gráfico de dispersão para mostrar a relação entre os bugs corrigidos e a produtividade
sns.scatterplot(x='Bugs Corrigidos', y='Produtividade', data=df)
plt.title('Relação entre Bugs Corrigidos e Produtividade')
plt.xlabel('Bugs Corrigidos')
plt.ylabel('Produtividade')
plt.show()
```

# Ao analisar as informações estatísticas básicas e os gráficos de dispersão, podemos obter insights sobre os dados.
* Por exemplo, podemos verificar se existe uma tendência de aumento da produtividade com o aumento das horas trabalhadas ou com o aumento do número de bugs corrigidos. Além disso, podemos identificar outliers ou valores atípicos que podem afetar nossas análises.

* Podemos também realizar análises estatísticas indutivas e dedutivas para obter mais informações sobre os dados. Por exemplo, podemos calcular o coeficiente de correlação de Pearson entre as variáveis para verificar se existe uma correlação linear entre elas.
* Além disso, podemos realizar testes de hipóteses para verificar se as diferenças observadas entre as médias são estatisticamente significativas.



```python
# Calculando o coeficiente de correlação de Pearson entre as horas trabalhadas e a produtividade
corr, p_value = stats.pearsonr(df['Horas Trabalhadas'], df['Produtividade'])
print('Coeficiente de Correlação de Pearson entre Horas Trabalhadas e Produtividade:', corr)

# Calculando o coeficiente de correlação de Pearson entre os bugs corrigidos e a produtividade
corr, p_value = stats.pearsonr(df['Bugs Corrigidos'], df['Produtividade'])
print('Coeficiente de Correlação de Pearson entre Bugs Corrigidos e Produtividade:', corr)
```
